## Data Cleaning

### Handling Missing Values

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print(f'Duplicates: {duplicates}')

# Calculate the percentage of missing values for each column
missing_percentage = data.isnull().mean() * 100

# Plot the missing data percentage for each column as a bar chart
plt.figure(figsize=(14, 8))
missing_percentage.sort_values(ascending=False).plot(kind='bar', color='skyblue', edgecolor='gray')
plt.title('Percentage of Missing Values by Column in EM-DAT Dataset', fontsize=16)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Missing Percentage (%)', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

### Filling Missing Values

```python
# Categorical columns: Fill with 'Unknown'
categorical_columns = [
    'Historic', 'Classification Key', 'Disaster Group', 'Disaster Subgroup', 'Disaster Type', 
    'Disaster Subtype', 'OFDA/BHA Response', 'Appeal', 'Declaration', 'Event Name', 
    'Origin', 'Associated Types', 'Admin Units', 'River Basin'
]
data[categorical_columns] = data[categorical_columns].fillna('Unknown')

# Geographical columns: 0 imputation
# data['Latitude'] = data['Latitude'].fillna(data['Latitude'].mean())
data['Latitude'] = data['Latitude'].fillna(0)
data['Longitude'] = data['Longitude'].fillna(0)

# Date columns: Forward fill for missing month and day in sequential events
date_columns = ['Start Month', 'Start Day', 'End Month', 'End Day']
data[date_columns] = data[date_columns].ffill()

# Numerical columns with zero for now
numerical_zero_fill = ['Total Deaths', 'No. Injured', 'No. Affected', 'No. Homeless', 'Total Affected']
data[numerical_zero_fill] = data[numerical_zero_fill].fillna(0)

# Economic loss columns 0 imputation for no impact, could do the median imputation too
economic_columns = [
    'Reconstruction Costs (\'000 US$)', 'Reconstruction Costs, Adjusted (\'000 US$)', 
    'Insured Damage (\'000 US$)', 'Insured Damage, Adjusted (\'000 US$)', 
    'Total Damage (\'000 US$)', 'Total Damage, Adjusted (\'000 US$)', 
    'AID Contribution (\'000 US$)'
]
for col in economic_columns:
    data[col] = data[col].fillna(0)
    #data[col] = data[col].fillna(data[col].median())

# Magnitude and CPI columns
data['Magnitude'] = data['Magnitude'].fillna(0)
data['Magnitude Scale'] = data['Magnitude Scale'].fillna('n/a')
# For the CPI, (Consumer Price Index), fill in with median. 
# Used to adjust US$ values for inflation
data['CPI'] = data['CPI'].fillna(data['CPI'].median()) 

# Entry Date and Last Update columns: Fill missing with latest known date or extraction date
latest_entry_date = data['Entry Date'].max()
latest_update_date = data['Last Update'].max()
data['Entry Date'] = data['Entry Date'].fillna(latest_entry_date)
data['Last Update'] = data['Last Update'].fillna(latest_update_date)

# Confirming that there are no missing values left
missing_after_imputation = data.isnull().sum()

# Display results
missing_after_imputation
data.info()
```

## Exploratory Data Analysis (EDA)

```python
# Load the dataset
data = pd.read_csv("./data/emdat_2020-2024_cleaned.csv")
# data.head()
```

```python
# Performing Simple EDA analysis

# Basic Statistical Summary for Numerical Columns
numerical_summary = data.describe()
print("Statistical Summary of Numerical Columns:")
print(numerical_summary)
```

```python
# Distribution Analysis for Key Numerical Columns
# Histograms for Total Deaths, Total Affected, Magnitude, Total Damage
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
data['Total Deaths'].plot(kind='hist', ax=axes[0, 0], bins=30, edgecolor='black', title='Total Deaths Distribution')
data['Total Affected'].plot(kind='hist', ax=axes[0, 1], bins=30, edgecolor='black', title='Total Affected Distribution')
data['Magnitude'].plot(kind='hist', ax=axes[1, 0], bins=30, edgecolor='black', title='Magnitude Distribution')
data['Total Damage (\'000 US$)'].plot(kind='hist', ax=axes[1, 1], bins=30, edgecolor='black', title='Total Damage Distribution')
plt.tight_layout()
plt.show()
```

```python
# Categorical Analysis - Bar plots for Disaster Type and Disaster Group
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
data['Disaster Type'].value_counts().plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black', title='Frequency of Disaster Types')
data['Disaster Group'].value_counts().plot(kind='bar', ax=axes[1], color='lightgreen', edgecolor='black', title='Frequency of Disaster Groups')
plt.tight_layout()
plt.show()
```

```python
# Time-Series Analysis - Trend of Total Deaths over Start Year
# Aggregating by 'Start Year' to observe trends

# Aggregate data by 'Start Year' to observe trends
yearly_data = data.groupby('Start Year').agg({
    'Total Deaths': 'sum',
    'Total Damage (\'000 US$)': 'sum'
})

# Create figure and axis objects with appropriate size
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot Total Deaths on the left y-axis with improved style
ax1.plot(
    yearly_data.index, 
    yearly_data['Total Deaths'], 
    color='royalblue', 
    marker='o', 
    linestyle='-', 
    linewidth=2, 
    label='Total Deaths'
)
ax1.set_ylabel('Total Deaths (count)', color='royalblue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='royalblue')

# Create a twin axis to plot Total Damage on the right y-axis
ax2 = ax1.twinx()
ax2.plot(
    yearly_data.index, 
    yearly_data['Total Damage (\'000 US$)'], 
    color='tomato', 
    marker='x', 
    linestyle='--', 
    linewidth=2, 
    label='Total Damage (000 US$)'
)
ax2.set_ylabel('Total Damage (000 US$)', color='tomato', fontsize=12)
ax2.tick_params(axis='y', labelcolor='tomato')

# Set x-axis label and main title
ax1.set_xlabel('Year', fontsize=12)
plt.title('Yearly Trend of Total Deaths and Economic Damage from Disasters (2000-2025)', fontsize=14, fontweight='bold')

# Adding a legend to indicate which line is which
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left', fontsize=10)

# Adding grid lines for better readability
ax1.grid(visible=True, which='major', linestyle='--', linewidth=0.5)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Correlation Analysis - Heatmap of Correlation between Numerical Columns

# Select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Generate the correlation matrix on the numeric subset
correlation_matrix = numeric_data.corr()

# Plot the heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

```python
# Create histograms with improved formatting
import math

# Calculate the number of numerical columns
num_columns = len(data.select_dtypes(include='number').columns)
num_rows = math.ceil(num_columns / 4)  # Adjust rows based on the number of columns

# Create subplots with dynamic row adjustment
fig, axes = plt.subplots(nrows=num_rows, ncols=4, figsize=(20, num_rows * 3))
fig.suptitle('Histograms of Numerical Features', fontsize=20)
fig.tight_layout(pad=3.0)  # Add space between plots for clarity

# Plot each column in a separate subplot
for i, column in enumerate(data.select_dtypes(include='number').columns):
    ax = axes[i // 4, i % 4]
    ax.hist(data[column].dropna(), bins=20, color='skyblue', edgecolor='black')
    ax.set_title(column)
    ax.tick_params(axis='x', rotation=45)

# Hide any unused subplots
for j in range(i + 1, num_rows * 4):
    fig.delaxes(axes[j // 4, j % 4])

plt.show()
```
