This notebook is responsible for performing exploratory data analysis on the newly-cleaned EM-DAT data from 1900-2024.

```python
# Import necessary libraries for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exploratory Data Analysis (EDA)

```python
# Load the dataset
data = pd.read_csv("./data/emdat_1900-2024_cleaned.csv")
data.head()
data.columns
```

```python
# Create a Frequency chart for Disaster Type
plt.figure(figsize=(12, 6))
sns.countplot(y='Disaster Type', data=data, order=data['Disaster Type'].value_counts().index)

plt.title('Frequency For each disaster')
plt.xlabel('Count')
plt.ylabel('Disaster Type')
# Use a logarithmic scale for plotting
plt.xscale('log')  
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

# Create a plot for Disaster Type x Total affected
plt.figure(figsize=(12,5))

sns.barplot(x='Region', y='No. Affected', data=data, ci=None)

# Format y-axis labels to avoid scientific notation (1e6)
ax = plt.gca()  # Get current axis
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.title('Number Affected Across Regions')
plt.xlabel('Regions')
plt.ylabel('No. Affected')
plt.show()
```
